```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_movie=pd.read_csv("IMDb Movies India.csv",encoding=("ISO-8859-1"),sep=",",engine='python')
df_movie.dropna(inplace=True)
df_movie.head()
```

```python
df_movie.shape
```

```python
df_movie.describe()
```

```python
df_movie.isna().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()

df_movie['Genre']= labelencoder.fit_transform(df_movie['Genre'])

df_movie.head()
```

```python
df_movie.isna().sum()
```

```python
df2=df_movie.drop(['Votes'],axis=1)
df2.head()
```

```python
df2.describe()
```

```python

df2.isna().sum()
```

```python
df_final=df2.dropna()
```

```python
df_final.shape   
```

```python

df_final.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style of seaborn plots
sns.set(style="whitegrid")

# Plotting a histogram of the 'Rating' column
plt.figure(figsize=(10, 6))
sns.histplot(df2['Rating'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()

# Plotting a count plot of the 'Genre' column
plt.figure(figsize=(10, 6))
sns.countplot(y='Genre', data=df2, order=df2['Genre'].value_counts().index[:10], palette='viridis')
plt.title('Top 10 Genres')
plt.xlabel('Count')
plt.ylabel('Genre')
plt.show()

# Plotting a bar plot of the top 10 Directors
plt.figure(figsize=(10, 6))
top_directors = df2['Director'].value_counts().nlargest(10)
sns.barplot(x=top_directors.values, y=top_directors.index, palette='rocket')
plt.title('Top 10 Directors')
plt.xlabel('Count')
plt.ylabel('Director')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style of seaborn plots
sns.set(style="whitegrid")

# Plotting the distribution of ratings over the years
plt.figure(figsize=(12, 6))
sns.boxplot(x='Year', y='Rating', data=df2, palette='viridis')
plt.title('Distribution of Ratings over the Years')
plt.xlabel('Year')
plt.ylabel('Rating')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()

# Plotting the relationship between duration and rating
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Duration', y='Rating', data=df2, alpha=0.5, color='skyblue')
plt.title('Relationship between Duration and Rating')
plt.xlabel('Duration (minutes)')
plt.ylabel('Rating')
plt.show()
```
